```python
from transformers import AutoTokenizer, pipeline
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

## Part 1: Input Sentences
Define sentences.

```python
sentences = [
    "I love this movie, it is amazing!",
    "This product is terrible and disappointing.",
    "AI is changing the world rapidly.",
    "The food was okay, not great.",
    "I would never recommend this to anyone.",
    "I am so happy because it's a warm and sunny day!",
    "The book was absolutely fantastic, a true masterpiece.",
    "I have an overwhelming feeling of unhappiness."
]

sentences
```

## Part 2: Tokenization Analysis
Use the tokenizers from BERT, GPT2, and LLaMA

```python
#load tokenizers
gpt2_tokenizer = AutoTokenizer.from_pretrained("gpt2")

bert_tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

llama_tokenizer = AutoTokenizer.from_pretrained("hf-internal-testing/llama-tokenizer")
```

```python
results = []

for sentence in sentences:
    bert_tokens = bert_tokenizer.tokenize(sentence)
    gpt2_tokens = gpt2_tokenizer.tokenize(sentence)
    llama_tokens = llama_tokenizer.tokenize(sentence)

    results.append({
        "Sentence": sentence,
        "BERT_count": len(bert_tokens),
        "BERT_tokens": " | ".join(bert_tokens),
        "GPT2_count": len(gpt2_tokens),
        "GPT2_tokens": " | ".join(gpt2_tokens),
        "LLaMA_count": len(llama_tokens),
        "LLaMA_tokens": " | ".join(llama_tokens),
    })

df = pd.DataFrame(results)
df.index = df.index + 1
df
```

### Questions
1. Which tokenizer produces the **most tokens on average**?
- LLaMA on averages produces the most tokens. It produced the definitive most in 5/7 standard trials, and the other 2/7 it tied for the most.
2. Which tokenizer uses **subword splitting**?
-  When handling rare or unseen words, like "unhappiness" in my own sentence 8, all three tokenizers use subword splitting. However, LLaMA is the most aggressive with its subword splitting, cutting down words like masterpiece, sunny, disappointing, and amazing. These are all pretty intelligent tokenizers though.
3. Briefly explain why tokenizers produce different outputs.
- Tokenizer output varies based on how a given tokenizer chooses to split text and represent information. BERT and LLaMA represent the contraction "'s" as the apostrophe and s characters separately, while in GPT2's vocabulary, there is a special vocabulary term for "'s". These three tokenizers handle spaces, punctuation, and rare words differently, which leads to variation in token boundaries and token counts.

## Part 3: Hugging Face Sentiment Analysis

```python
classifier = pipeline("sentiment-analysis")
```

```python
hf_results = []

for sentence in sentences:
    result = classifier(sentence)[0]
    hf_results.append({
        "Sentence": sentence,
        "HF_Label": result["label"],
        "HF_Score": result["score"]
    })

df_hf = pd.DataFrame(hf_results)
df_hf.index = df_hf.index + 1
df_hf
```
